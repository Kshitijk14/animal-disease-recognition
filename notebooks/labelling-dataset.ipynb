```python
import os
import shutil
import pandas as pd
```

## For Train folder

```python
# Define paths
dataset_folder = r"cattle diseases multiclass\train"
csv_file = os.path.join(dataset_folder, "_classes.csv")  # CSV file path
destination_folder = os.path.join(dataset_folder, "sorted_images")  # Folder to store sorted images

# Create the destination folder if it doesn't exist
os.makedirs(destination_folder, exist_ok=True)

# Read the CSV file
df = pd.read_csv(csv_file)
```

```python
df
```

```python
# Loop through each row to find the label column with value 1
for index, row in df.iterrows():
    filename = row['filename']
    
    # Identify the label (column name where value is 1)
    label = df.columns[1:][row[1:] == 1].values[0].strip()  # Ignore 'filename' column
    
    # Create a directory for each label if it doesn't exist
    label_path = os.path.join(destination_folder, label)
    os.makedirs(label_path, exist_ok=True)

    # Source and destination paths for the image
    src_path = os.path.join(dataset_folder, filename)
    dest_path = os.path.join(label_path, filename)

    # Move the image to the corresponding folder
    if os.path.exists(src_path):
        shutil.move(src_path, dest_path)
        print(f"Moved: {filename} -> {label_path}")
    else:
        print(f"File not found: {filename}")

print("Sorting complete!")
```
